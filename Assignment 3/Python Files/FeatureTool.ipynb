```python
import featuretools as ft
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("loan.csv")
```

```python
train.head()
```

```python
# saving missing values in a variable
a = train.isnull().sum()/len(train)*100
# saving column names in a variable
variables = train.columns
variable = [ ]
for i in range(0,74):
    if a[i]>=50:   #setting the threshold as 50%
        variable.append(variables[i])
```

```python
df = train.drop(variable,axis=1)
```

```python
df.head()
```

```python
df = df.drop(['member_id','delinq_2yrs','earliest_cr_line','inq_last_6mths','open_acc','pub_rec'],axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['emp_title'].fillna("missing", inplace = True)
```

```python
df['emp_length'].fillna(0, inplace = True)
```

```python
df['title'].fillna("missing", inplace = True)
```

```python
df['revol_util'].fillna(df['revol_util'].mean(), inplace = True)
df['annual_inc'].fillna(df['annual_inc'].mean(), inplace = True)
```

```python
df['total_acc'].fillna(0, inplace = True)
```

```python
df['last_pymnt_d'].fillna("", inplace = True)
```

```python
df['next_pymnt_d'].fillna("", inplace = True)
```

```python
df['last_credit_pull_d'].fillna("", inplace = True)
```

```python
df['collections_12_mths_ex_med'].fillna(0, inplace = True)
df['acc_now_delinq'].fillna(0, inplace = True)
```

```python
df['tot_coll_amt'].fillna(0, inplace = True)
df['tot_cur_bal'].fillna(0, inplace = True)
df['total_rev_hi_lim'].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("Clean_Dataset.csv")
```

```python
es = ft.EntitySet(id = 'Data Set')
```

```python
es.entity_from_dataframe(entity_id = 'Lending Club', dataframe = df, index = 'id')
```

```python
feature_matrix, feature_names = ft.dfs(entityset=es, 
target_entity = 'Lending Club',                                                                       
max_depth = 1, 
verbose = 1, 
n_jobs = 3)
```
